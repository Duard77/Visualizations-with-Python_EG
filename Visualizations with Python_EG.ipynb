```python
""""
Use the below datasets create any four visualization summarizing the datasets. You can use either Matplotlib, Seaborn or any other python based visualization package. Be creative!. Here are some example visualizations: https://blog.hubspot.com/marketing/types-of-graphs-for-data-visualization. Remember to label and have legends for your graphs. Add comments to describe what you are trying to graph
"""

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer/breast-cancer.data",
                names = ["Class", "Age", "Menopause", "Tumor size", "Inv-nodes", "Node-caps", "Deg-Malig", "Breast", "Breast quad", "Irridiat"])
```

```python
#Line graph: Total Count of Breast Cancer cases vs. Age range
Line = df.groupby('Age').count()  #This groups the data by Breast and Breast quadrant for the number of cases
Line['Total'] = Line['Class']   #creates new variable 'Total' in groupby df out of count total of column 'Class'
plt.title("Total Count of Breast Cancer cases by Age range")   #Label the Graph
plt.ylabel("Total Cases")   #Label the y-axis
Line['Total'].plot()  #Since 'Age' is the index because of the groupby, we can plot 'Total' against it
```

```python
#Histogram graph: Distribution of Breast Cancer by Breast Quadrant count
Hist = df.dropna()
Hist = Hist.reset_index(drop=True)
Hist = Hist.groupby(['Breast quad']).count()  #This groups the data by Breast and Breast quadrant for the number of cases
Hist['Total'] = Hist['Class']
Hist['Frequency'] = Hist['Total']/285
#print(Hist)
plt.title("Distribution of Breast Cancer Frequency by Breast Quadrant Counts")   #Label the Graph
plt.ylabel("Frequency")#Label the y-axis
plt.xlabel("Breast Quadrant Counts")
plt.hist(Hist.Total)
plt.show()
```

```python
#Pie Chart: Tumor size distribution in Breast Cancer cases
Pie = df.dropna()
Pie = Pie.reset_index(drop=True)
p1 = Pie.loc[Pie['Tumor size'] == "0-4"].count()[0]  #This groups the data by Tumor size for the number of cases
p2 = Pie.loc[Pie['Tumor size'] == "5-9"].count()[0]
p3 = Pie.loc[Pie['Tumor size'] == "10-14"].count()[0]
p4 = Pie.loc[Pie['Tumor size'] == "15-19"].count()[0]
p5 = Pie.loc[Pie['Tumor size'] == "20-24"].count()[0]
p6 = Pie.loc[Pie['Tumor size'] == "25-29"].count()[0]
p7 = Pie.loc[Pie['Tumor size'] == "30-34"].count()[0]
p8 = Pie.loc[Pie['Tumor size'] == "35-39"].count()[0]
p9 = Pie.loc[Pie['Tumor size'] == "40-44"].count()[0]
p10 = Pie.loc[Pie['Tumor size'] == "45-49"].count()[0]
p11 = Pie.loc[Pie['Tumor size'] == "50-54"].count()[0]
print(p1)

#Pie = Pie.groupby(['Tumor size']).count()   #Checking the counts
#print(Pie)

plt.title("Tumor Size distribution in Breast Cancer Cases")   #Label the Graph
labels = ["0-4", "5-9", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40-44", "45-49", "50-54"]   #labels
plt.pie([p1, p2, p3, p4, p5, p6, p7, p8, p9, p10, p11], labels=labels, autopct='%.1f %%')    #format the pie pieces
plt.legend(loc = 'upper right', fontsize = 'xx-small', title = "Tumor Size:")   #format the legend
plt.show()
```

```python
#Box and Whisker: Distribution of Degree of Malignancy by Age comparison in Breast Cancer Cases
BW = df.dropna()   #Drop NaN
BW = BW.reset_index(drop=True)   #Resets the index
A1 = BW.loc[BW.Age == "20-29"]['Deg-Malig']   #creates tuples out of data groupings of age and deg-malig
A2 = BW.loc[BW.Age == "30-39"]['Deg-Malig']
A3 = BW.loc[BW.Age == "40-49"]['Deg-Malig']
A4 = BW.loc[BW.Age == "50-59"]['Deg-Malig']
A5 = BW.loc[BW.Age == "60-69"]['Deg-Malig']
A6 = BW.loc[BW.Age == "70-79"]['Deg-Malig']
labels = "20-29", "30-39", "40-49", "50-59", "60-69", "70-79"   #creates labels
plt.title("Distribution of Degree of Malignancy by Age Comparison in Breast Cancer Cases")   #Label the Graph
plt.ylabel("Degree of Malignancy")#Label the y-axis
plt.xlabel("Age")  #Label x-axis
plt.boxplot([A1, A2, A3, A4, A5, A6], labels=labels)   #creates box and whisker
plt.show()
```
